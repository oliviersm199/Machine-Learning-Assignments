# Question 1

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as pyplot
from matplotlib.pyplot import scatter

def q1_read_datasets(file_path):
    dataset = pd.read_csv(file_path,header=None)
    dataset = dataset[[0,1]]
    dataset.columns = ["X","Y"]
    return dataset

def q1_convert_to_numpy(dataframe):
    return dataframe['X'].as_matrix(), dataframe['Y'].as_matrix()



def q1_set_model_degree(X,degree):
    vectors = []
    for i in range(degree + 1):
        if i == 0:
            vectors.append(np.ones(len(X)))
        else:
            vectors.append(np.power(X,i))
    return np.asarray(vectors).transpose()


def q1_fit_model(X,Y):
    x_transpose = X.transpose()
    x_t_inverse = np.linalg.inv(np.matmul(x_transpose, X))
    return np.matmul(np.matmul(np.linalg.inv(np.matmul(x_transpose, X)),x_transpose),Y)
    
def print_out_polynomial(weights):
    return_string = ""
    for i,w in enumerate(weights):
        if i == 0:
            return_string = return_string + "%.5f + "%(w)
        else:
            return_string = return_string + "%.5fx^%d +" % (w,i)
    return return_string[:-1]


def get_mean_square_error(X,Y,W):
    error = np.subtract(Y, np.matmul(X,W))
    error_transpose = error.transpose()
    return np.matmul(error_transpose, error)[0][0] / W.shape[0]


ds1_train = q1_read_datasets("Datasets/Dataset_1_train.csv")
ds1_train_x, ds1_train_y = q1_convert_to_numpy(ds1_train)


XTRAIN = q1_set_model_degree(ds1_train_x,20)
YTRAIN = ds1_train_y.reshape(len(ds1_train_y),1)
WTRAIN = q1_fit_model(XTRAIN,YTRAIN)



x_graph = np.arange(-1.00,1.00001,0.00001)
scatter(x_graph, np.polynomial.polynomial.polyval(x_graph,WTRAIN),s=0.5)
scatter(ds1_train_x, ds1_train_y)

pyplot.xlabel('Input')
pyplot.ylabel('Output')
pyplot.ylim((-20,20))
pyplot.title('Q1 - 1')

pyplot.show()

pyplot.close()

   
print("Train Error: %.5f "%(get_mean_square_error(XTRAIN,YTRAIN,WTRAIN)))

ds1_test = q1_read_datasets("Datasets/Dataset_1_test.csv")
ds1_test_x, ds1_test_y  = q1_convert_to_numpy(ds1_test)

XTEST = q1_set_model_degree(ds1_test_x,20)
YTEST = ds1_test_y.reshape(len(ds1_test_y),1)

print("Test Error: %.5f "%(get_mean_square_error(XTEST,YTEST,WTRAIN)))



```

Quality of the fit seems to be quite good for the majority of the function but around positive and negative 1 it does not estimate the data correctly. Also the training error while small at 15 does not generalize well for the test set and we get a much higher test error. This suggests that an n = 20 model is overfitting the data.



```python
def q1_fit_model_l2(X,Y,lambda_var):
    x_transpose = X.transpose()
    mult_x_x_transpose = np.matmul(x_transpose,X)
    identity_matrix_size = mult_x_x_transpose.shape[0]
    lambda_identity_matrix = lambda_var * np.identity(identity_matrix_size)
    return np.matmul(np.matmul(np.linalg.inv(np.add(mult_x_x_transpose, lambda_identity_matrix)),x_transpose),Y)

# for different values of lambda  
errors = []

for i in np.arange(0.0, 1.001, 0.001):
    WTRAINL2 = q1_fit_model_l2(XTRAIN, YTRAIN,i)
    train_error = get_mean_square_error(XTRAIN,YTRAIN,WTRAINL2)
    test_error = get_mean_square_error(XTEST,YTEST,WTRAINL2)
    errors.append((train_error, test_error,i))
    #print("Train Error: %.5f\tValidation Error: %.5f\tLambda: %.5f"%(train_error,test_error,i))

lambda_vars = [err[-1] for err in errors]
trainset_vars = [err[0] for err in errors]
testset_vars = [err[1] for err in errors]
scatter(lambda_vars,trainset_vars)
scatter(lambda_vars,testset_vars)
pyplot.show()
pyplot.close()
```

We clearly get a huge improvement by introducing a regularization term as seen by the significant difference in training and testing error above a lambda value of 0. However it's difficult to see what value of lambda gives the best performance. So the first value of lambda = 0 is removed since it is difficult to visualize. 



```python
lambda_vars_no_zero = lambda_vars[1:]
trainset_vars_no_zero = trainset_vars[1:]
testset_vars_no_zero = testset_vars[1:]

scatter(lambda_vars_no_zero,trainset_vars_no_zero)
scatter(lambda_vars_no_zero,testset_vars_no_zero)
pyplot.show()
pyplot.close()
```

The best value of lambda was determined by looking at the percentage improvement in generalization (i.e. our performance on the test set) as we increase the value of lambda. This was then plotted to see if there was a significant drop off. 

```python
previous_error = errors[0]

percentage_improvements = []
for error in errors[1:]:
    percentage_improvement = abs((error[1] - previous_error[1]) / previous_error[1]) * 100
    previous_error = error
    percentage_improvements.append(percentage_improvement)
    
scatter(lambda_vars_no_zero,percentage_improvements)
pyplot.show()
    
```

The first introduction of a lambda turn led to an almost 80% improvement in test accuracy. However after this improvement, subsequent variations in lambda only lead to incremental improvements in test accuracy. To see if there was any sharp dropoff, the first improvement was discarded. 

```python
scatter(lambda_vars_no_zero[1:],percentage_improvements[1:])
pyplot.show()
```

As you can see the percentage improvement in the test accuracy gradually declines as we increase the value of lambda. This means that at some point, increasing lambda becomes insignificant in terms of significantly improving the test results. This point was determined to be after 0.1 and this was decided to be the best value for lambda.

```python
final_improvements = zip(lambda_vars,percentage_improvements,trainset_vars,testset_vars)
final_lambda = 0
for lambda_var,improvement, train_error,test_error in final_improvements:
    if improvement < 0.1:
        break
    final_lambda = (lambda_var,train_error,test_error)

final_lambda
    
```

The best value for lambda was determined to be 0.010999999999999999 with a training error of 20.985 and a validation error of 25.739.

Subsequently it was tested to see if the lambda chosen could help us approximate or guess the degree of the source polynomial. The weights of the model were thus examined:


```python
WTRAINL2 = q1_fit_model_l2(XTRAIN, YTRAIN,0.010999999999999999)
scatter(XTRAIN, np.polynomial.polynomial.polyval(XTRAIN,WTRAINL2))
scatter(ds1_train_x, ds1_train_y)
pyplot.show()
```

```python
scatter(XTEST,np.polynomial.polynomial.polyval(XTEST,WTRAINL2))
scatter(ds1_test_x,ds1_test_y)
pyplot.show()
```

The quality of the fit is much better than the previous model which had a very poor generalization at the ends. This seems to follow the data much more smoothly.

From the visualization I would infer that the model used to generate the test data has a degree of 2.

```python
train_data = q1_read_datasets('./Datasets/Dataset_2_train.csv')
test_data = q1_read_datasets('./Datasets/Dataset_2_test.csv')
validate_data = q1_read_datasets('./Datasets/Dataset_2_valid.csv')

q2_x_train, q2_y_train = q1_convert_to_numpy(train_data)
q2_x_test, q2_y_test = q1_convert_to_numpy(test_data)
q2_x_validate, q2_y_validate = q1_convert_to_numpy(validate_data)


```

```python
import random

# Linear Model
Q2XTRAIN = q1_set_model_degree(q2_x_train,1)
Q2YTRAIN = q2_y_train.reshape(len(q2_y_train),1)
Q2XVALIDATE = q1_set_model_degree(q2_x_validate,1)
Q2YVALIDATE = q2_y_validate.reshape(len(q2_y_validate),1)

WSTART = np.asarray([[1.],[1.]])
alpha = 1e-6
epsilon = 0.0001


def online_gradient_descent(X, Y, WSTART, alpha, epsilon,XVALIDATE, YVALIDATE):
    MSE_LIST = []
    W_LIST = []
    WCURRENT = WSTART
    while True:
        WNEXT = WCURRENT
        training_set = list(zip(X,Y))
        random.shuffle(training_set)
        for sample_x, sample_y in training_set:
            sample_x_vec = sample_x.reshape(2,1)
            sample_x_vec_transpose = sample_x_vec.transpose()
            sample_y_vec = sample_y.reshape(1,1)
            gradient = np.multiply(2,np.subtract(np.matmul(np.matmul(sample_x_vec,sample_x_vec_transpose),WNEXT),np.matmul(sample_x_vec,sample_y_vec)))
            WNEXT = WNEXT - alpha * gradient
            
        TRAIN_MSE = get_mean_square_error(X,Y,WNEXT)
        VALIDATE_MSE = get_mean_square_error(XVALIDATE, YVALIDATE, WNEXT)
        MSE_LIST.append((TRAIN_MSE, VALIDATE_MSE,WCURRENT))
            
        
        if np.linalg.norm(np.subtract(WNEXT,WCURRENT)) < epsilon:
            break
        
        WCURRENT = WNEXT
        
    return WCURRENT, MSE_LIST

weights, MSE_LIST = online_gradient_descent(Q2XTRAIN, Q2YTRAIN, WSTART, alpha, epsilon, Q2XVALIDATE, Q2YVALIDATE)

scatter(q2_x_train, q2_y_train)
scatter(q2_x_train,np.polynomial.polynomial.polyval(q2_x_train,weights))
pyplot.show()
```

```python
epoch = list(range(len(MSE_LIST)))

scatter(epoch, [error[0] for error in MSE_LIST], label="Training Error")
scatter(epoch, [error[1] for error in MSE_LIST], label ="Test Error")


pyplot.legend()
pyplot.title("MSE Error for Q2")
pyplot.xlabel("Epoch")
pyplot.ylabel("MSE Error")

pyplot.show()
```

```python
#
# On this step I'm testing various step sizes and graphing them
#

alpha_list = [1e-5,1e-4,1e-3, 0.00175 ,0.0025, 0.0050]


MSE_LIST_ALPHA = []

for alpha in alpha_list:
    weights, MSE_LIST = online_gradient_descent(Q2XTRAIN, Q2YTRAIN, WSTART, alpha, epsilon, Q2XVALIDATE, Q2YVALIDATE)
    MSE = get_mean_square_error(Q2XVALIDATE, Q2YVALIDATE, weights)
    print(alpha, MSE, len(MSE_LIST))
    MSE_LIST_ALPHA.append((alpha,MSE))

    
alphas = [elem[0] for elem in MSE_LIST_ALPHA]
mse_errors = [elem[1] for elem in MSE_LIST_ALPHA]


scatter(alphas, mse_errors)
pyplot.show()
    
    
    


```

```python
def q3_get_dataset(data, headers):
    with open(headers,'r') as labels_reader:
        labels = [line.strip() for line in labels_reader]
        return pd.read_csv(data,names = labels)

q3_df = q3_get_dataset('./Datasets/communities.data.txt','./Datasets/labels.txt')

columns_to_fill_na = list(q3_df[q3_df.columns[5:]].columns)


# Replacing with mean 
mean_not_available = []
for column in columns_to_fill_na:
    temp_column = q3_df[column].apply(lambda x : np.nan if x == "?" else float(x))
    q3_df.drop(labels=[column], axis="columns", inplace=True)
    q3_df[column] = temp_column
    mean = q3_df[column].mean()
    q3_df[column] = q3_df[column].fillna(mean)

q3_df.to_csv('./Dataset_Q3')
q3_df.head()
```

The process of replacing a missing value with the mean is preffered over an alternative like deleting the columns with missing data as it does not reduce the effective sample size. The samples with missing data points may follow a pattern that is not random and deleting these would cause the new sample set to not be representitive of the original. Furthermore the mean is fairly simple to calculate and substitute in but reduces the variability and variance of the data, thus introducing some bias. [1] 



```python
def split_into_n_folds(df, n_folds):
    fold_numbers = set(q3_df['fold'])
    folds_per_group = len(fold_numbers) / n_folds
    fold_count = 0
    fold_list = []
    for fold in range(n_folds):
        fold_temp = q3_df[(q3_df.fold < (fold_count + folds_per_group)) & (q3_df.fold > fold_count)]
        fold_list.append(fold_temp)
        fold_count = fold_count + 1 
    return fold_list

def q3_to_array(df):
    y = "ViolentCrimesPerPop"
    y_temp = df[y].as_matrix()
    y_temp = y_temp.reshape(len(y_temp),1)
    x_temp = np.asarray([df[column].as_matrix() for column in df if column != y]).transpose()
    return x_temp, y_temp

    
    
def get_training_folds(train_fold_index,folds):
    return pd.concat([fold for j, fold in enumerate(folds) if j != train_fold_index])
   
    
def cross_validate(folds, question_num, fitting_function,*args):
    MSE_LIST = [] 
    for i, validation_fold in enumerate(folds): 
        # Get the Training/Validation Datasets and save them as a csv.
        training_fold = get_training_folds(i, folds)
        training_fold.to_csv("Q%d-CandC-train%d.csv"%(question_num ,i))
        validation_fold.to_csv("Q%d-CandC-test%d.csv"%(question_num ,i))
        
        # Remove non-normalized, continous variables
        training_fold = training_fold[training_fold.columns[5:]]
        validation_fold = validation_fold[validation_fold.columns[5:]]
        
        # Convert to numpy matrices 
        validateX, validateY = q3_to_array(validation_fold)
        trainX, trainY = q3_to_array(training_fold)
        
        # Train Model on Training Fold (80%) and Get MSE on Validation Fold (20%)
        weights = fitting_function(trainX,trainY, *args)
        MSE = get_mean_square_error(validateX,validateY,weights)
        MSE_LIST.append({"MSE":MSE,"validation_fold":i,"weights":weights})
        
        
    return MSE_LIST
    

def get_average_mse(mse_list):
    return np.average((experiment['MSE'] for experiment in mse_list))


fold_list = split_into_n_folds(q3_df,5)
mse_list = cross_validate(fold_list,2, q1_fit_model)
cross_validation_error = get_average_mse(mse_list)

print("The Average Cross Validation Error is %.5f"%(cross_validation_error))
print("The parameters for the five models were:")
for iteration, test in enumerate(mse_list):
    print("Fold %d"%(iteration))
    for parameter_name, weight in zip(q3_df.columns[5:],test['weights']):
        print("Parameter:%s\tWeight:%s"%(parameter_name, str(weight[0])))
    print("\n\n")
```

```python
lambdas = np.arange(0,1.1,0.1)
cross_validation_errors = [] 
for i in np.arange(0,1.1,0.1):
    mse_list = cross_validate(fold_list, 2, q1_fit_model_l2, i)
    cross_validation_error = get_average_mse(mse_list)
    cross_validation_errors.append(cross_validation_error)

scatter(lambdas, cross_validation_errors)

pyplot.show()


    


```

The MSE does not depreciate significantly with L2 regularization. It seems to plateau around 0.035 so a regularization constant lambda of 0.1 was chosen since there isn't significant improvement after this point in the cross validation error. From this regularization, the coefficients of the new model help in feature selection. The importance of a variable in the linear regression is proportional to the magnitude of the coefficient. This is due to  all terms in the dataset being preprocessed and regularized to have a value between 0 and 1. 

```python

def get_most_important_features(mse_list,df,non_predictive_features):
    feature_coefficient_dict = {}
    for mse in mse_list:
        for feature in zip(df[df.columns.difference(non_predictive_features)],list(mse['weights'].transpose()[0])):
            if feature[0] in feature_coefficient_dict:
                feature_coefficient_dict[feature[0]].append(feature[1])
            else:
                feature_coefficient_dict[feature[0]] = [feature[1]]

    # Take the average importance not regularized. This will account for the values that vary 
    # between being positive and negative, thus having no predictive power over different folds 
    average_coefficient_dict = {}
    for feature in feature_coefficient_dict:
        average_coefficient_dict[feature] = np.average(feature_coefficient_dict[feature])

    return sorted([{"feature":feature, "coefficient":average_coefficient_dict[feature]} for feature in average_coefficient_dict], reverse=True ,key = lambda x: abs(x["coefficient"]))

def get_key_features(feature_list, threshhold):
    key_features = filter(lambda x : abs(x["coefficient"]) > threshhold,feature_list)
    return [feature['feature'] for feature in key_features]


train_df = q3_df
i = 0 
while i < 2:
    non_predictive_features = ["state","county","community","communityname","fold","ViolentCrimesPerPop"]
    temp_to_sort = get_most_important_features(mse_list, train_df, non_predictive_features)
    key_features = get_key_features(temp_to_sort, 0.1)
    keep_features = non_predictive_features + key_features
    fold_list = split_into_n_folds(train_df[keep_features],5)
    mse_list = cross_validate(fold_list,3, q1_fit_model_l2, 0.1)
    mse = get_average_mse(mse_list)
    train_df = train_df[keep_features]
    print(keep_features)
    i = i + 1 
    

    
    
    
    
    
```
